```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

#import keras
```

```python
class MultiLayerPerceptron:
    # 
    #
    #
    #The architecture of the neural network will be defined by the python tuple "arqt", of which the number of 
    #entries gives the number of layers (and thus the number of synapses matrices will be "len(arqt)-1"). 
    #Last but not least, the value of each of the entries in "arqt" defines the number of nodes in the corresponding layer
    #
    # 
    #
    
 
    
    def __init__(self, arqt, option, reg = 2, l2=0.001, epochs=1000, eta=0.02, 
                 alpha=0.001, decrease_eta=0.0001, shuffle=True, 
                 minibatches=100,check_grad=False, number_of_checks=5, epsilon=1e-5):
        
        self.w = []                  # w will be a list of matrices where we´ll store the weights (synapses) matrices
        self.z = []                  # To store the various Z = WA+b 
        self.a = []                  # List to store the a = f(z), being a[0] = X
        self.option = option             # Defines which non-linear function will be used in each layer
        
        self.n_output = arqt[-1]
        self.n_features = arqt[0]
        self.reg = 2
        self.l2 = l2
        self.epochs = epochs
        self.eta = eta
        self.alpha = alpha
        self.decrease_eta = decrease_eta
        self.shuffle = shuffle
        self.minibatches = minibatches
        self.check_grad = check_grad
        self.number_of_checks = number_of_checks
        self.epsilon = epsilon
        
        # The next loop inicialize weights of all the synapses matrices, based on the information given by the arqt tuple 
        # The weight matrices already consider the biases
        for m,n in zip(arqt[:-1],arqt[1:]):
            matrix = self._random_matrix_(n,m)
            self.w.append(matrix)
        
     
    # The role of the next method is to initialize the weights with numbers in [-1,1]
    # Parameters n and m are integers representing the dimentions of the weight matrix
    def _random_matrix_(self,n,m):
        return np.random.randn(n,m+1)/np.sqrt(m+1)
    
    # In case we are interested 
    def _onehot(self,y, k):
        print (k, y.shape[0])
        onevsrest = np.zeros((k, y.shape[0]))
        for indice, classe in enumerate(y):
            onevsrest[classe, indice] = 1.0
        return onevsrest
        
    # Here is the place to choose which non-linearity we´re going to use to active our unities 
    def _non_lin(self,z,option=1):
        
        # logistic function (instance of sigmoid function)
        # classes associated with the output of this function are 0 and 1
        if option == 1:
            return (1.0 + np.exp(-1*z))**(-1.0)
        
        # Hyperbolic tangent (also a sigmoid): similar to the logistic function, but output classes -1 and 1
        elif option == 2:
            return tanh(z)
        
        # Softmax: turns scores into probabilities
        # Specially useful for multiclass classification and aimed to be used in the output layer
        elif option == 3:
            return np.exp(z)/np.sum(np.exp(z),axis=0)
        
        # Linear 
        elif option == 4:
            return z
        
        else: 
            raise AttributeError('`option` must be 1, 2, 3 or 4 only')
    
    # This method gives the gradients we´re going to use 
    # Bearing in mind that the softmax will be used only in the last layer, its gradient won´t be needed
    def _derivatives(self,z,option=1):
        
        if option == 1:
            a = self._non_lin(z,1)   
            return a*(1.0-a)
        
        if option == 2:
            a = self._non_lin(z,2)   
            return 1-(a**2.0)
        
    # Adding the bias unit  
    def _insert_bias(self,X,current_layer):
        
        X_bias = np.copy(X)
        if current_layer == 0:
            ones = np.ones((X.shape[0],1))
            #print "dim(ones) = {0} and dim(X) = {1}".format(ones.shape,X.shape)
            X_bias = np.hstack([ones,X])
            
        else:
            ones = np.ones((X.shape[1]))
            X_bias = np.vstack([ones,X])
            
        return X_bias
    
    # Forward propagation
    def _f_prop(self,X,w,option):
        
        z = []
        a = []
        
        for i in np.arange(len(w)):
            #print "i", i
            if i == 0:
                a.append(self._insert_bias(X,i))
                z.append(w[i].dot(a[i].T))
                #print "this is {0} and this is {1}".format(z,option[i])
                a.append(self._non_lin(z[i],self.option[i]))
                #print a
                
            else:
                a[i] = self._insert_bias(a[i],i)
                #print a
                z.append(w[i].dot(a[i]))
                a.append(self._non_lin(z[i],option[i]))
                #print a
                
                
        return a,z
    
    # Will be needed for gradient check. the parameter k will be given by the gradient_check function
    def _f_prop_for_grad_check(self, X, w, k, epsilon, ep_matrices, minus = False):
        
        z = []
        a = []
        W = w[:]
        
        if minus == False:
            W[k] = W[k] + ep_matrices[k]
        
        else:
            W[k] = W[k] - ep_matrices[k]
            
        a, z = self._f_prop(X, W, self.option)
                
        return a,z,W
        
    
    # Regularization term 
    def _reg(self, w, reg = 2):
        tudo = 0
        # L2 (ridge)
        if reg == 2:
            for i in np.arange(len(w)):
                tudo = tudo + np.sum(w[i][:, 1:])**2
            
            return (self.l2/2.0) * tudo 
            
        # L1 (lasso)
        elif reg == 1:
            for i in np.arange(len(w)):
                tudo = tudo + np.abs(w[i][:,1:]).sum()
            
            return (self.l1/2.0) * tudo
        
        else:
            raise AttributeError('`reg` must be 1 or 2')
            
    # Cost function: we will be using (generalized) Cross entropy.  
    def _cost_function(self, w, target, output_layer, option, reg = 2):
        
        if option[-1] == 1:
            cost = np.sum(-(target * (np.log(output_layer)))-((1 - target) * np.log(1 - output_layer)))
            
        if option[-1] == 3:
            cost = np.sum(-(target * (np.log(output_layer))))
            
        reg_term = self._reg(w, reg)
        cost = cost + reg_term
        
        return cost
    
    # A very naive adaptive learning rate choice 
    def _adaptive_learning_rate(self, n):
        
        self.eta = self.eta/(1 + n*self.decrease_eta)
    
    # Shuffling the data to make sure we´re gonna get different batches each time around
    # Important: y here means the matrix returned by _one_hot
    def _shuffle(self, X, y):
        
        aux = np.random.permutation(np.arange(X.shape[0]))
        #print "dim(shuffle_X) = {0} and dim(shuffle_Y = {1})".format(X.shape,y.shape)
        if self.n_output > 1:
            X, y = X[aux], y[:, aux]
        else:
            X, y = X[aux], y[aux]
        return X, y
    
    # Creating the minibatches 
    def _minibatches(self,X,minibatches):
        
        return np.array_split(np.arange(X.shape[0]),self.minibatches)
    
    # Backpropagation  
    def _back_prop(self, a, z, y_label, w, option):
    
        gradients = []     
        
        for i in np.arange(len(w)):
            
            j = -(i+1)
            if  j == -1:   # Output layer
                delta = a[j] - y_label
                gradients.append(np.dot(delta,a[j-1].T))
            # All the other layers   
            elif i < (len(w)-1):
                z_bias = self._insert_bias(z[j], j)
                #print "\n ELIF z_bias = {0}".format(z_bias)
                delta = w[j+1].T.dot(delta)*self._derivatives(z_bias,option[j])
                delta = delta[1:, :]
                #print "\n ELIF i = {0} and j = {1}".format(i,j)
                gradients.append(np.dot(delta,a[j-1].T))
            else:
                z_bias = self._insert_bias(z[j], j)
                #print "\n ELSE z_bias = {0}".format(z_bias)
                delta = w[j+1].T.dot(delta)*self._derivatives(z_bias,option[j])
                delta = delta[1:, :]
                #print "\n ELSE i = {0} and j = {1}".format(i,j)
                gradients.append(np.dot(delta,a[j-1]))
        
        gradients.reverse() 
        
        # Don´t forget to regularize
        for i in np.arange(len(w)):
            #print "\n gradients[i][:, 1:].shape is {0} and w[i][:, 1:].shape is {1} for j = {2}".format(gradients[i][:, 1:].shape,w[i][:, 1:].shape,i)
            gradients[i][:, 1:] += (w[i][:, 1:] * self.l2)
            
        return gradients    
    # The next function generates a list of zero matrices with approppriate shapes to store 
    # .. the previus delta, at all times. 
    def _last_delta_w(self, w):
         
        prev_w = []
        
        for i in np.arange(len(w)):
            prev_w.append(np.zeros(w[i].shape))
            
        return prev_w
            
            
    def _gradient_checking(self, X, y_hot, w, epsilon, gradients,batch):
        """ Apply gradient checking (for debugging only)

        Returns
        ---------
        relative_error : float
          Relative error between the numerically
          approximated gradients and the backpropagated gradients.

        """
        
        numeric_gradients = []
        epsilon_matrices = []
        num_grads = [] 
        grad = []
        
        for i in np.arange(len(w)):
            epsilon_matrices.append(np.zeros(np.shape(w[i])))
            numeric_gradients.append(np.zeros(np.shape(w[i])))
    
        for k in np.arange(len(w)):
            for i in np.arange(w[k].shape[0]):
                for j in np.arange(w[k].shape[1]):
                    epsilon_matrices[k][i,j] = epsilon
                    a, z, w_k_plus = self._f_prop_for_grad_check(X, w, k, epsilon, epsilon_matrices, minus = False)
                    cost_plus = self._cost_function(w_k_plus, y_hot, a[-1][:, batch], self.option, reg = 2)
                    a, z, w_k_minus = self._f_prop_for_grad_check(X, w, k, epsilon, epsilon_matrices, minus = True)
                    cost_minus = self._cost_function(w_k_minus, y_hot, a[-1][:, batch], self.option, reg = 2)
                    numeric_gradients[k][i, j] = (cost_plus - cost_minus) / (2 * epsilon)
                    epsilon_matrices[k][i,j] = 0
                    
        num_grads = numeric_gradients[0].flatten()
        #print "\nnum_grads inicial", num_grads
        grad = gradients[0].flatten()
        #print "\ngrad inicial", grad
        for i in np.arange(1,len(w)):
            num_grads = np.hstack((num_grads,numeric_gradients[i].flatten()))
            #print "\n num_grads: {0} passo{1}".format(num_grads,i)
            grad = np.hstack((grad,gradients[i].flatten()))
            #print "\ngrad: {0} passo{1}".format(grad,i)
        
        norm1 = np.linalg.norm(num_grads - grad)
        norm2 = np.linalg.norm(num_grads)
        norm3 = np.linalg.norm(grad)
        relative_error = norm1 / (norm2 + norm3)
        return relative_error
    
    def predict(self, X):
        #print "\nw está aqui",self.w
        a, z = self._f_prop(X,self.w,self.option)
        #print "AAAAAAAAAAAAAAAA = {0} and dim(A)={1}".format(a[-1],a[-1].shape)
        if self.n_output > 1:                            # When you have more than 2 classes
            y_pred = np.argmax(a[-1], axis=0)
            
        else:
            y_pred = a[-1]
        
        return y_pred
            
    def fit(self, X, y, print_progress=False):
        #print "shape of X is {0} and of Y is{1}".format(X.shape,y.shape)
        self.cost_ = []
        X_data, y_data = X.copy(), y.copy()
        if self.n_output > 1:                            # When you have more than 2 classes
            y_hot = self._onehot(y, self.n_output)
        else: 
            y_hot = y_data
        #print "\ny_hot = {0} and DIM(y_hot) = {1}\n".format(y_hot,y_hot.shape)
        last_delta_w = self._last_delta_w(self.w)
        
        for i in np.arange(self.epochs):
            print "EPOCA",i
            self._adaptive_learning_rate(i)
            
            if self.shuffle:
                X_data, y_hot = self._shuffle(X_data, y_hot)
                #print "shuffle_X = {0} and shuffle_Y = {1}".format(X_data,y_hot)
            mini = self._minibatches(y_hot,self.minibatches)
            #print "\nmini\n",mini
            for j,batch in enumerate(mini):
                #print "X_data[batch] = {0}, y_hot[batch] = {1} and batch = {2}".format(X_data[batch],y_hot[batch],batch)
                a, z = self._f_prop(X_data[batch],self.w,self.option)
                #print "AAAAAAAAAAAAAAAA = {0} and dim(A)={1}".format(a[-1],a[-1].shape)
                if self.n_output > 1:
                    cost = self._cost_function(self.w, y_hot[:, batch], a[-1], self.option, reg = 2)
                    self.cost_.append(cost)
                    gradients = self._back_prop(a, z, y_hot[:, batch], self.w, self.option)
                else:
                    cost = self._cost_function(self.w, y_hot[batch], a[-1], self.option, reg = 2)
                    print "\nCOST COST COST\n",cost
                    self.cost_.append(cost)
                    gradients = self._back_prop(a, z, y_hot[batch], self.w, self.option)
                    
                if ((self.check_grad == True)and(i <= self.number_of_checks)):
                    if self.n_output > 1:
                        grad_diff = self._gradient_checking(X_data[batch], y_hot[:, batch], self.w, self.epsilon, gradients,batch)
                    else:
                        grad_diff = self._gradient_checking(X_data[batch], y_hot[batch], self.w, self.epsilon, gradients,batch)
                    
                    if grad_diff <= 1e-7:
                        print('Ok: %s' % grad_diff)
                    elif grad_diff <= 1e-4:
                        print('Warning: %s' % grad_diff)
                    else:
                        print('PROBLEM: %s' % grad_diff)
                        
                for k in np.arange(len(self.w)):
                    delta_w = self.eta * gradients[k] 
                    self.w[k] = self.w[k] - (delta_w + self.alpha*last_delta_w[k])
                    last_delta_w[k] = delta_w
                    
        return self
                    
   
```

```python
from sklearn.model_selection import train_test_split
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target
X_digits_train, X_digits_test, y_digits_train, y_digits_test = train_test_split(X_digits, y_digits, stratify=y_digits ,test_size = 0.2, random_state =10)
```

```python
nn = MultiLayerPerceptron((64,200,10),(1,1,3)).fit(X_digits_train, y_digits_train, print_progress=False)
```

```python
plt.plot(range(len(nn.cost_)), nn.cost_)
plt.ylim([0, 4])
plt.ylabel('Cost')
plt.xlabel('Epochs')
plt.tight_layout()
# plt.savefig('./figures/cost.png', dpi=300)
plt.show()

y_train_pred = nn.predict(X_digits_train)
```

```python
got_right = (y_digits_train == y_train_pred).sum()
acc = float(got_right)/X_digits_train.shape[0]
print X_digits_test.shape[0]
print got_right
print "Accuracy, in the training set, for this run is {0}%".format(acc*100)
```

```python
y_test_pred = nn.predict(X_digits_test)
got_right_test = (y_digits_test == y_test_pred).sum()
acc = float(got_right_test)/X_digits_test.shape[0]
print X_digits_test.shape[0]
print got_right_test
print "Test accuracy for this run is {0}%".format(acc*100)
```
